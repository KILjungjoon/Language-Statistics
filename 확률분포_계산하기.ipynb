* (1)번 계산

```python
from scipy.stats import norm

# 정규분포 정의의
mu = 30
sigma = 4

# norm 클래스 안의 누적분포함수 계산에 쓰이는 cdf 메소드 이용
# cdf는 주어진 값보다 같거나 작은 값을 가질 확률을 계산해 준다
prob = norm.cdf(32, mu, sigma) - norm.cdf(26, mu, sigma)

# 출력
print("Probability of P(26<=X<=32) = {:.4f}".format(prob))
```

* (3)번 계산

```python
# 확률식만 바꿔주면 된다다
prob = norm.cdf(34, mu, sigma)
print("Probability of P(X<=34) = {:.4f}".format(prob))
```
