```python
from scipy.stats import norm

#  
mu = 30
sigma = 4

# norm      cdf  
# cdf         
prob = norm.cdf(32, mu, sigma) - norm.cdf(26, mu, sigma)

# 
print("Probability of P(26<=X<=32) = {:.4f}".format(prob))
```
